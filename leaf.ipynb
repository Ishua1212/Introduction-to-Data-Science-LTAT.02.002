```python
# IMPORT LIBRARYS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# IMPORTING DATA
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
sample_submission = pd.read_csv('sample_submission.csv')
```

```python
# SET VARIABLES
seed=1
np.random.seed(seed=seed)
```

```python
# COPY
train_copy = train.copy()
```

```python
# OVERVIEW
arr = {
    "Null": train_copy.isnull().sum().sum(),
    "Duplicates": train_copy.duplicated().sum(),
    "Types": train_copy.dtypes.value_counts().to_dict(),
}

for key, value in arr.items():
    print(f"{key}: {value}")

print(f"Shape: {train_copy.shape}")
print(f"Uniques: {train_copy['species'].value_counts().size}")
```

```python
# PREPARE DATA FOR TRAINING
train_copy['species_id'], species_names = pd.factorize(train_copy['species'])

x = train_copy.drop(['id', 'species', 'species_id'], axis=1)
y = train_copy['species_id']

x.head()
```

```python
# TRAIN AND VALIDATION SPLIT
x_train, x_val, y_train, y_val = train_test_split(x, y, random_state=5, test_size=0.2)
```

```python
# RFC
rf = RandomForestClassifier(n_estimators=1000, max_depth=10, random_state=seed)
rf.fit(x_train, y_train)
rf_accuracy = accuracy_score(y_val, rf.predict(x_val))
print(f"RFC accuracy: {rf_accuracy}")
```

```python
# KNNC
knn = KNeighborsClassifier(n_neighbors=1, metric='euclidean')
knn.fit(x_train, y_train)
knn_accuracy = accuracy_score(y_val, knn.predict(x_val))
print(f"KNN accuracy: {knn_accuracy}")
```
