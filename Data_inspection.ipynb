```python
import glob
from matplotlib import rcParams
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

rcParams['figure.figsize'] = 15, 20
plt.gray()

# get the path/directory
folder_dir = 'images'
image_paths = glob.iglob(f'{folder_dir}/*')
images = []
for path in image_paths:
    images.append(mpimg.imread(path))


```

```python
subplot_index = 1
# iterate over images
for image in images:
    plt.subplot(10, 10, subplot_index)
    plt.grid(False) # Remove grid
    plt.imshow(image)
    subplot_index +=1
    if subplot_index > 100: break
    
```
