```python
# Load images
import glob
from matplotlib import rcParams
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

rcParams['figure.figsize'] = 15, 20
plt.gray()

# Get the path/directory
folder_dir = '../Data/images'
image_paths = glob.iglob(f'{folder_dir}/*')
images = []
for path in image_paths:
    images.append(mpimg.imread(path))
```

```python
# Shows 100 images
subplot_index = 1
# Iterate over images
for image in images:
    plt.subplot(10, 10, subplot_index)
    plt.grid(False) # Remove grid
    plt.imshow(image)
    subplot_index +=1
    if subplot_index > 100: break
```
